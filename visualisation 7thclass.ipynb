# line plot

```python
#import matplotlib.pyplot as plt
from matplotlib import pyplot as plt
x=[2,3,6,5]
y=[5,6,7,5]
plt.plot(x,y,marker='o')
plt.show()
```

```python
x=[2,2,5,5,2]
y=[2,5,5,2,2]
plt.plot(x,y,marker='*')
plt.xlim(0,10) # putting the limit
plt.ylim(0,10) # putting the limit
plt.title("rectangle") # title of the graph
plt.xlabel("x_label")# putting the x label
plt.ylabel("y_label")# putting the y label
plt.show()
```

# styles

```python
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
x=[1,2,3,4,5]
y=[5,4,3,2,1]
x1=[6,5,4,3,2]#[2,3,4,5,6]
y1=[2,3,4,5,6]#[6,5,4,3,2]
plt.plot(x,y,'r',label='lineone',linewidth=5)
plt.plot(x1,y1,'c',label='linetwo',linewidth=5)
plt.title("linegraph")
plt.xlabel('x-label')
plt.ylabel('y_label')
plt.legend()
plt.grid(True,color='r')
plt.show()
```

# bargraph: plot the categorical variable or a combination of continuous variable

```python
x=[1,3,5,7,9]
x1=[2,4,6,8]
y=[10,15,18,19,20]
y1=[11,16,19,20]
plt.bar(x,y,color='r')
plt.bar(x1,y1,color='g')
plt.show()
```

# programming language bar graph

```python
labels=["javascript","java","pyhton","c#"]
usage=[69.5,45.3,38.8,34.4]

y_positions =range(len(labels))

plt.bar(y_positions,usage,color="k")
plt.xticks(y_positions,labels)
plt.ylabel("Usage(%)")
plt.title("programming language usage")
plt.show()
```

# scatter plot: to create relationship between continuous variable

```python
temp=[30,32,33,28.5,35,29,29]
ice_cream_counts=[100,115,115,75,125,79,89]

plt.scatter(temp,ice_cream_counts,color='k',s=500,marker='o')
plt.title("temp vs ice_cream_counts")
plt.xlabel("Temperature")
plt.ylabel("Solid ice creams counts")
plt.show()
```

```python
x1=[4]
y1=[4]
plt.bar(x1,y1,color='#3d130c',width=1)
x2=[5]
y2=[4]
plt.bar(x2,y2,color='#f5faf5',width=1)
x3=[6]
y3=[4]
plt.bar(x3,y3,color='#062906',width=1)
x=[5]
y=[5]
plt.scatter(x,y,color='#2a33db',marker='o',s=1000)
plt.xlim(0,10)
plt.ylim(0,10)
plt.show()
```

# pieplot:  use for distribution

```python
slices=[7,2,2,13]
activities=['sleeping','eatng','working','playing']
cols=['c','m','r','b']

plt.pie(slices,
        labels=activities,
        colors=cols,
        startangle=90,
        shadow=True,
        explode=(0,0.5,0,0),
        autopct='%1.1f%%')
plt.title("pie_plot")
plt.show()
```
