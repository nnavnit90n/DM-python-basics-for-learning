```python
from matplotlib import pyplot as plt
days=[1,2,3,4,5]

sleeping=[7,8,6,11,7]
eating=[2,3,4,3,2]
working=[7,8,7,2,2]
playing=[8,5,7,8,13]
plt.plot([],[],color='m',label='sleeping',linewidth=5)
plt.plot([],[],color='c',label='eating',linewidth=5)
plt.plot([],[],color='r',label='working',linewidth=5)
plt.plot([],[],color='k',label='playing',linewidth=5)
plt.stackplot(days,sleeping,eating,working,playing,colors=['m','c','r','k'])
         
plt.xlabel('x')
plt.ylabel('y')
plt.title('stack plot')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(0,3*np.pi,0.1)
print(x)
y=np.sin(x)
plt.plot(x,y,color='r')
plt.grid(True,color='g')
plt.xlabel("range")
plt.ylabel("frequency")
plt.plot
plt.show()
```

# histogram plot: distribute the data in frequency and bins

```python
from matplotlib import pyplot as plt
numb=[0.1,0.5,1,1.5,2,4,5.5,6,8,9]
plt.hist(numb,bins=11)
plt.xlabel("numbers")
plt.ylabel("frequency")
plt.show()
```

# boxplot:
it is use to plot a combination of categorical and continuous variables.
spread of data and detecting the outliers.

```python
value=[1,2,5,6,6,7,7,8,8,8,9,10,21]
plt.boxplot(value)
plt.yticks(range(1,22))
plt.ylabel("value")
plt.show()
```
